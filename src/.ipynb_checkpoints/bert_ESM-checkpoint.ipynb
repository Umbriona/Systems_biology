```python
import csv
import pickle

# Read the CSV file
csv_file = 'Influenza_training_dataset.csv'


bert_dict_list = []
groundtruth_dict_list = []
# Open the CSV file 
with open(csv_file, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        p_identifier = row[0]
        p = row[1]
        c_identifier = row[2]
        c = row[3]
        if p != c:
            dict_likelihood = model.predict_likelihood(p,c)
            bert_dict_list.append(dict_likelihood)
        
             # Compare sequences character by character
            mutations = [(i, parent_char, child_char) for i, (parent_char, child_char) in enumerate(zip(p, c), start=1) if parent_char != child_char]
            # Calculate the position of mutations and update the statistics
            mutation_groundtruth = {}
            for position, parent_char, child_char in mutations:
                mutation_key = f"{parent_char}->{child_char}"
                mutation_groundtruth[position] = mutation_key
            groundtruth_dict_list.append(mutation_groundtruth)


            
        

file_path_groundtruth = 'groundtruth_dictionaries.pkl'
file_path_bert = 'bert_dictionaries.pkl'

# Save the dictionaries using pickle
with open(file_path_groundtruth, 'wb') as file:
    pickle.dump(groundtruth_dict_list, file)

with open(file_path_bert, 'wb') as file:
    pickle.dump(bert_dict_list, file)
```

```python
with open(file_path_groundtruth, 'rb') as file:
    groundtruth_dict_list = pickle.load(file)
with open(file_path_bert, 'rb') as file:
    bert_dict_list = pickle.load(file)

parent_probability_list = []
child_probability_list = []
for i in range(len(groundtruth_dict_list)): 
    groundtruth_mutation = groundtruth_dict_list[i]
    bert_mutation = bert_dict_list[i]
    for key, value in groundtruth_mutation.items():
        position = int(key)
        p_char = value[0]
        c_char = value[-1]
        try: 
            parent_probability = bert_mutation[position-1][p_char]
            parent_probability_list.append(parent_probability)
        except: 
            break
        try: 
            child_probability = bert_mutation[position-1][c_char]
            child_probability_list.append(child_probability)
        except: 
            break
    
```

```python
mutation_count = 0
first_probability_p = 0
second_probability_p = 0
third_probability_p = 0
first_probability_c = 0
second_probability_c = 0
third_probability_c = 0
for i in range(len(groundtruth_dict_list)): 
    groundtruth_mutation = groundtruth_dict_list[i]
    bert_mutation = bert_dict_list[i]
    mutation_count += len(groundtruth_mutation)
    for key, value in groundtruth_mutation.items():
        position = int(key)
        p_char = value[0]
        c_char = value[-1]
        keys_list = list(bert_mutation[position-1].keys())
        if p_char == keys_list[0]:
            first_probability_p += 1
        elif p_char == keys_list[1]:
            second_probability_p += 1
        elif p_char == keys_list[2]:
            third_probability_p += 1
        if c_char == keys_list[0]:
            first_probability_c += 1
        elif c_char == keys_list[1]:
            second_probability_c += 1
        elif c_char == keys_list[2]:
            third_probability_c += 1
print(f'Number of mutation: {mutation_count}')
print(f'There are {first_probability_p} parent amino acid that ranked the most probable from bert, {second_probability_p} second probable, {third_probability_p} third probable')
print(f'There are {first_probability_c} child amino acid that ranked the most probable from bert, {second_probability_c} second probable, {third_probability_c} third probable')
        
```
